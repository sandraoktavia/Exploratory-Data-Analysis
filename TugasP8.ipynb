```python
plt.figure(figsize=(15, 5))
numeric_cols = data.select_dtypes(include='number').columns
for i, col in enumerate(numeric_cols[:3]):
    plt.subplot(1, 3, i+1)
    sns.histplot(data[col], kde=True, bins=30)
    plt.title(f'Distribusi Data: {col}')
plt.tight_layout()
plt.show()
```

```python
print("\nKorelasi Antar Kolom Numerik:")
pairs = [
    ('Age', 'Tumor Size'),
    ('Grade', 'Stage'),
    ('Diagnosis Year', 'Survival Months')
]

for col1, col2 in pairs:
    if col1 in data.columns and col2 in data.columns:
        corr_value = data[[col1, col2]].corr().iloc[0, 1]
        print(f'Korelasi antara {col1} dan {col2}: {corr_value:.2f}')
    else:
        print(f"Kolom '{col1}' atau '{col2}' tidak ditemukan di dataset.")
```

```python
# Diagram Batang
if 'Class' in data.columns:
    plt.figure(figsize=(8, 5))
    data['Class'].value_counts().plot(kind='bar', color='skyblue')
    plt.title('Frekuensi Tiap Kelas Tumor')
    plt.xlabel('Kelas')
    plt.ylabel('Jumlah')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python


# Pie Chart
if 'Class' in data.columns:
    plt.figure(figsize=(6, 6))
    data['Class'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)
    plt.title('Proporsi Kelas Tumor')
    plt.ylabel('')
    plt.tight_layout()
    plt.show()
```

```python
if len(numeric_cols) >= 2:
    x1, y1 = numeric_cols[0], numeric_cols[1]
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=x1, y=y1, data=data)
    plt.title(f'Scatterplot: {x1} vs {y1}')
    plt.xlabel(x1)
    plt.ylabel(y1)
    plt.tight_layout()
    plt.show()
else:
    print("Dataset tidak punya minimal dua kolom numerik untuk scatterplot pertama.")

if len(numeric_cols) >= 4:
    x2, y2 = numeric_cols[2], numeric_cols[3]
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=x2, y=y2, data=data)
    plt.title(f'Scatterplot: {x2} vs {y2}')
    plt.xlabel(x2)
    plt.ylabel(y2)
    plt.tight_layout()
    plt.show()
else:
    print("Dataset tidak punya cukup kolom numerik untuk scatterplot kedua.")
```
